```
import numpy as np
```

```
X = np.array(([2, 9], [1, 5], [3, 6]), dtype=float)
y = np.array(([92], [86], [89]), dtype=float)

X = X/np.amax(X, axis=0)
y = y/100
```

```
class Neural_Network(object):
  def __init__(self):
    self.input_size = 2
    self.output_size = 1
    self.hidden_size = 3
    
    self.input_to_hidden = np.random.randn(self.input_size, self.hidden_size)
    self.hidden_to_output = np.random.randn(self.hidden_size, self.output_size)
  
  def sigmoid(self, s):
    # activation function 
    return 1/(1+np.exp(-s))
  
  def sigmoidPrime(self, s):
    #derivative of sigmoid
    return s * (1 - s)
  
  
  def forward(self, X):
    self.hidden_input = np.dot(X, self.input_to_hidden)
    self.hidden_output = self.sigmoid(self.hidden_input)
    
    self.final_input = np.dot(self.hidden_output, self.hidden_to_output)
    self.final_ouput = self.sigmoid(self.final_input)
    
    return self.final_ouput
  
  def backward(self, X, y, o):
    
    self.output_error = y - o
    self.output_delta = self.output_error*self.sigmoidPrime(o)
    
    self.hidden_error = self.output_delta.dot(self.hidden_to_output.T)
    self.hidden_delta = self.hidden_error*self.sigmoidPrime(self.hidden_output)
    
    self.input_to_hidden += X.T.dot(self.hidden_delta)
    self.hidden_to_output += self.hidden_output.T.dot(self.output_delta)
    
    
  def train (self, X, y):
    o = self.forward(X)
    self.backward(X, y, o)
    
    
NN = Neural_Network()
for i in range(1000): # trains the NN 1,000 times
  if(i%10 == 0):
    print("Input: \n" + str(X)) 
    print("Actual Output: \n" + str(y))
    print("Predicted Output: \n" + str(NN.forward(X)))
    print("Loss: \n" + str(np.mean(np.square(y - NN.forward(X))))) # mean sum squared loss
    print("\n")
  NN.train(X, y)
```
