```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
from PIL import Image
import urllib


%matplotlib inline
```

```
url="https://raw.githubusercontent.com/tajain07/neural-network-playground/master/curved_lane.jpg"


f = urllib.request.urlopen(url)

a = plt.imread(f, 0)
plt.imshow(a)
#plt.show()
```

#### Convert the image to grayscale

```
gray = cv2.cvtColor(a, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```
# Horizontal kernel

# 3x3 array for edge detection
horizontal_filter = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])

## TODO: Create and apply a Sobel x operator


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(gray, -1, horizontal_filter)

plt.imshow(filtered_image, cmap='gray')
```

```
# Vertical kernel

# 3x3 array for edge detection
vertical_filter = np.array([[ 2, 0, -2], 
                   [ 2, 0, -2], 
                   [ 2, 0, -2]])
                            
                      
vertical_filter = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])


# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(gray, -1, vertical_filter)

plt.imshow(filtered_image, cmap='gray')
```

```
blur_filter = np.array([[ 0.0625, 0.125, 0.0625], 
                   [ 0.125, 0.25, 0.125], 
                   [ 0.0625, 0.125, 0.0625]])

vertical_filter = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])

# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(gray, -1, blur_filter)
filtered_image = cv2.filter2D(filtered_image, -1, vertical_filter)

plt.imshow(filtered_image, cmap='gray')
```
